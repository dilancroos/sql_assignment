```python
import requests
import json

# Get data from API
def get_data_from_api(api_url):
    try:
        response = requests.get(api_url)
        response.raise_for_status()  # Check for any request errors

        data = response.json()  # Parse the JSON data from the response

        return data
    except requests.exceptions.RequestException as e:
        print(f"Error: {e}")
        return None

# Save data to JSON file
def save_data_to_json(data, file_path):
    try:
        with open(file_path, 'w') as json_file:
            json.dump(data, json_file, indent=4)
        print(f"Data saved to {file_path}")
    except Exception as e:
        print(f"Error saving data to JSON file: {e}")
```

```python
# Main program to get the data from the API and save it to a JSON file
api_url = "https://swapi.dev/api/"  
output_file = "dataLinks.json"  # Specify the path and filename for the JSON file

data = get_data_from_api(api_url)

if data:
    save_data_to_json(data, output_file)
```

```python
# INNER JOIN
# people with a species (mentioned) where person name starts with a 'D'

pd.read_sql_query(
    '''
        SELECT
            p.name AS people_name,
            s.name AS species_name
        FROM
            people AS p
        INNER JOIN
            species AS s ON p.species_id = s.species_id
        WHERE
            p.name LIKE 'D%'

    ''',
    con)
```

```python
# LEFT JOIN
# Top 5 species with the most number people

pd.read_sql_query(
    '''
        SELECT
            sp.name,
            COUNT(*) AS num_people
        FROM
            species AS sp
        LEFT JOIN
            people AS pe ON pe.species_id = sp.species_id
        GROUP BY
            sp.name
        ORDER BY
            num_people DESC
        LIMIT 5
    ''',
    con)
```

```python
# GROUP BY
# Top 5 planets with the most number of people living on them (excluding unknown)

pd.read_sql_query(
    '''
        SELECT
            pl.name,
            COUNT(*) AS num_people
        FROM
            planets AS pl
        INNER JOIN
            people AS pe ON pe.homeworld_id = pl.planet_id
        WHERE 
            pl.name != 'unknown'
        GROUP BY
            pl.name
        ORDER BY
            num_people DESC
        LIMIT 5
    ''',
    con)
```

```python
# HAVING
# Get the pilots name and the number of starships he/she has piloted and the starship having a cargo_capacity greater than 100000 and the pilot has piloted more than 1 starship

pd.read_sql_query(
    '''
        SELECT
            p.name AS pilot_name,
            COUNT(*) AS num_starships_piloted
        FROM
            people AS p
        INNER JOIN
            people_starships AS ps ON ps.people_id = p.people_id
        INNER JOIN
            starships AS s ON s.starships_id = ps.starships_id
        GROUP BY
            p.name
        HAVING
            num_starships_piloted > 1 AND s.cargo_capacity > 100000
        ORDER BY
            num_starships_piloted DESC

    ''',
    con)
```

```python
# CTE
# Get the number of starships and vehicles having a cargo_capacity greater than 10000
pd.read_sql_query(
    '''
        WITH starships_cap AS (
            SELECT
                'starship' AS type,
                COUNT(*) AS num_starships_vehicles
            FROM
                starships AS s
            WHERE
                s.cargo_capacity > 10000
        ),
        vehicles_cap AS (
            SELECT
                'vehicle' AS type,
                COUNT(*) AS num_starships_vehicles
            FROM
                vehicles AS v
            WHERE
                v.cargo_capacity > 10000
        )

        SELECT
            *
        FROM
            starships_cap
        
        UNION
        
        SELECT
            *
        FROM
            vehicles_cap
    ''',
    con)
```
